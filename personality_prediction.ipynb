**Label Encoding and One-Hot Encoding**

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# List of categorical columns
object_lst = ['Gender','Education', 'Age Group', 'Interest']
encoder = LabelEncoder()
dict1 = {}

# Label encoding for categorical features
for i in object_lst:
    dict1[i] = LabelEncoder()
    df[i] = dict1[i].fit_transform(df[i])

# One-Hot encoding for the 'Personality' column
onehot = OneHotEncoder(sparse_output=False, drop='first')
personality_encoded = onehot.fit_transform(df[['Personality']])

# Convert the one-hot encoded 'Personality' into a DataFrame
personality_df = pd.DataFrame(personality_encoded, columns=[f'Personality_{i}' for i in range(personality_encoded.shape[1])])
```

**Prepare Data for Training**

```python
# Select features for training
x = df[['Gender', 'Education', 'Introversion Score', 'Sensing Score', 'Thinking Score',
        'Judging Score', 'Interest', 'Age Group']]

# The target variable is the one-hot encoded 'Personality'
y = personality_df

# Convert 'y' into single class labels (argmax to get class index)
y = np.argmax(y.values, axis=1)

# Apply Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()

# Fit the scaler on training data (x) and transform it
x_scaled = minmax.fit_transform(x)
```

```python
# Loop through the dictionary to get the mappings for each encoder
inverse_mappings = {}

for column, encoder in dict1.items():
    # Inverse transform the encoded labels
    inverse_mapping = dict(zip(encoder.transform(encoder.classes_), encoder.classes_))
    inverse_mappings[column] = inverse_mapping

# Display the inverse mappings
for column, inverse_mapping in inverse_mappings.items():
    print(f"'{column}':")
    print(inverse_mapping)
    print()
```

**Train-Test Split**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x_scaled,y,random_state=42,test_size=0.3)
```

**Feature Selection using RFE**

```python
# Feature Selection using Recursive Feature Elimination (RFE)
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# RFE for selecting the top 10 features
rfe = RFE(estimator=RandomForestClassifier(), n_features_to_select=10)
x_train_rfe = rfe.fit_transform(x_train, y_train)  # Fit on training data
x_test_rfe = rfe.transform(x_test)  # Transform the test data based on fitted RFE model
```

**Model Training and Evaluation**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score

# Define models for training and evaluation
models = [
    LogisticRegression(max_iter=1000),  # Logistic Regression
    KNeighborsClassifier(),             # K-Nearest Neighbors
    SVC(),                              # Support Vector Classifier
    GaussianNB(),                       # Naive Bayes
    DecisionTreeClassifier()            # Decision Tree
]

# Train and evaluate each model
for model in models:
    model.fit(x_train_rfe, y_train)  # Train the model
    y_pred = model.predict(x_test_rfe)  # Predict on test data

    print(model)
    print(f'Accuracy: {round((accuracy_score(y_test, y_pred)) * 100)}%')  # Print accuracy
    print(classification_report(y_test, y_pred))  # Print detailed classification report
    print()
```

**Hyperparameter Tuning with GridSearchCV**

```python
from sklearn.model_selection import GridSearchCV

# Define models for hyperparameter tuning
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    # "SVC": SVC(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier()
}

# Hyperparameter grids for tuning
param_grids = {
    "Logistic Regression": {
        'C': [0.1, 1, 10],
        'solver': ['liblinear', 'lbfgs'],
        'max_iter': [100, 200, 300]
    },
    "K-Nearest Neighbors": {
        'n_neighbors': [3, 5, 7, 9],
        'weights': ['uniform', 'distance'],
        'algorithm': ['auto', 'ball_tree', 'kd_tree']
    },
    # "SVC": {
    #     'C': [0.1, 1, 10],
    #     'kernel': ['linear', 'rbf', 'poly'],
    #     'gamma': ['scale', 'auto'],
    #     'degree': [3, 5]
    # },
    "Naive Bayes": {
        'var_smoothing': [1e-9, 1e-8, 1e-7]
    },
    "Decision Tree": {
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'criterion': ['gini', 'entropy']
    }
}

# Perform grid search for hyperparameter tuning
for model_name, model in models.items():
    print(f"Optimizing {model_name}...")
    grid_search = GridSearchCV(
        model,
        param_grids[model_name],
        cv=5,
        n_jobs=-1,
        verbose=2,
        scoring='accuracy'  # Scoring metric
    )
    grid_search.fit(x_train_rfe, y_train)  # Fit the grid search on RFE-transformed training data

    # Output best parameters and score
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    print(f"Best cross-validated score for {model_name}: {round(grid_search.best_score_ * 100, 2)}%")
    print()

    # Get the best model
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(x_test_rfe)  # Predict on test data

    # Evaluate the best model performance
    print(f"Test Accuracy for {model_name}: {round(accuracy_score(y_test, y_pred) * 100, 2)}%")
    print(classification_report(y_test, y_pred))
    print()
```

**Using the Best Model to Predict Personality**

```python
# Access the best model after grid search
best_model = grid_search.best_estimator_

# Print the best model
print("Best Model after Tuning:", best_model)
```

```python
# Now you can use this `best_model` for training and prediction.
best_model.fit(x_train_rfe, y_train)
```

```python
def predict_personality():

    # Input numerical features from the user
    Gender = input("Enter Gender (Male/Female): ").strip()
    Age_Group = input("Enter Age Group (Young/Middle-Aged/Old): ").strip()
    Education = int(input("Enter Education level (as integer, e.g., 1 for High School, 2 for Bachelor's, etc.): "))
    Introversion_Score = float(input("Enter Introversion Score: "))
    Sensing_Score = float(input("Enter Sensing Score: "))
    Thinking_Score = float(input("Enter Thinking Score: "))
    Judging_Score = float(input("Enter Judging Score: "))
    Interest = input("Enter Interest category (as string, e.g., 'Arts', 'Science', etc.): ").strip()

    # Process categorical features
    gender_encoded = dict1["Gender"].transform([Gender])
    age_group_encoded = dict1["Age Group"].transform([Age_Group])

    try:
        interest_encoded = dict1["Interest"].transform([Interest])
    except ValueError:
        print(f"Unknown category '{Interest}' encountered for Interest. Using default encoding (0).")
        interest_encoded = [0]  # Default encoding for unseen categories

    # Combine numerical and encoded categorical features
    x_num = np.array([[gender_encoded[0], Education, Introversion_Score, Sensing_Score, Thinking_Score, Judging_Score, age_group_encoded[0]]])
    x_combined = np.hstack([x_num, interest_encoded.reshape(1, -1)])

    # Apply feature scaling using the already fitted scaler
    x_scaled = minmax.transform(x_combined)

    # Predict using the best model
    prediction = best_model.predict(x_scaled)

    # Convert the predicted label to one-hot encoding
    one_hot_prediction = np.zeros(personality_encoded.shape[1])
    one_hot_prediction[prediction] = 1

    # Decode the one-hot encoded personality output
    predicted_personality = onehot.inverse_transform(one_hot_prediction.reshape(1, -1))

    print(f"Predicted Personality: {predicted_personality[0][0]}")

predict_personality()
```
